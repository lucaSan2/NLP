```python
import tensorflow
```

```python
import pandas as pd
import numpy as np
```

```python
import keras
```

```python
from keras import Sequential
```

```python
from keras.layers import Dense
```

```python
wages = pd.read_csv('/content/hourly_wages.csv')
wages.head()
```

```python
wages.describe()
```

```python
target = wages['wage_per_hour']
```

```python
predictors = wages.drop('wage_per_hour', axis=1).values
```

```python
model = Sequential()
```

```python
model.add(Dense(50, activation='relu', input_shape=(predictors.shape[1],)))
```

```python
model.add(Dense(1, activation='relu'))
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model.compile(optimizer='RMSprop', loss='mean_squared_error')
```

```python
model.summary()
```

```python
stopper= EarlyStopping(patience=3)
```

```python
model.fit(predictors, target, validation_split=0.3, epochs=20, callbacks=[stopper])
```

```python
predictions = model.predict(predictors)
```

```python
predictions[:3]
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_absolute_error as mse
```

```python
mse(target, predictions)
```

```python
titanic = pd.read_csv('/content/titanic.csv')
titanic.head()
```

```python
dummies_class = pd.get_dummies(titanic['pclass'], drop_first=True, prefix='class')
```

```python
dummies_sib = pd.get_dummies(titanic['sibsp'], drop_first=True, prefix='sib')
```

```python
dummies_parch =pd.get_dummies(titanic['parch'], drop_first=True, prefix='parch')
```

```python
cat_features= pd.concat([dummies_class, dummies_sib, dummies_parch], axis=1)
cat_features.head()
```

```python
predictors2 = pd.concat([titanic, cat_features], axis=1).drop(['pclass', 'parch', 'sibsp'], axis=1)
predictors2.head()
```

```python
predictors2.describe()
```

```python
predictors2 = np.array(predictors2, np.float32)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
target2 = to_categorical(titanic['survived'])
```

```python
model = Sequential()
```

```python
model.add(Dense(50, activation='sigmoid', input_shape=(predictors2.shape[1],)))
```

```python
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
stopper= EarlyStopping(patience=3)
```

```python
model.fit(predictors2, target2, epochs=20, validation_split=0.3, callbacks=[stopper])
```
