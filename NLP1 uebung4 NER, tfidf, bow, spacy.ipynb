```python
wiki = pd.read_csv('../Datasets/people_wiki.csv')
wiki
```

```python
token_wiki = [word_tokenize(text.lower()) for text in wiki['text'][:100]]
```

```python
tags = [nltk.pos_tag(text) for text in token_wiki]
```

```python
chunks = nltk.ne_chunk_sents(tags)
```

```python
for sent in chunks:
    for chunk in chunks:
            if hasattr(chunk, 'label'):
                print(chunk)
```

```python
for sent in chunks:
    for chunk in chunks:
        print(chunk.label())
```

```python
ner_categories = defaultdict(int)
```

```python
for sent in chunks:
    for chunk in sent:
        ner_categories[chunk.label()] +=1
```

```python
movies = pd.read_csv('../Datasets/wiki_movie_plots_deduped.csv')[:50]
```

```python
movies2  = movies[movies['Genre'] != 'unknown']
```

```python
movies2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
movies2['Genre2'] = le.fit_transform(movies2['Genre'])
```

```python
movies2.head()
```

```python
 X = movies2['Plot']
```

```python
y= movies2['Genre2']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
c_vec = CountVectorizer(stop_words='english')
```

```python
c_train = c_vec.fit_transform(X_train)
```

```python
c_test = c_vec.transform(X_test)
```

```python
c_vec.get_feature_names()[:10]
```

```python
c_train.A[:15]
```

```python
count_df = pd.DataFrame(c_train.toarray(), columns=c_vec.get_feature_names())
```
