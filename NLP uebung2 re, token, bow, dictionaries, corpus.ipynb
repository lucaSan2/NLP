```python
def PowerFunction(artikel, lem):
    token = [text for text in word_tokenize(artikel.lower()) if text not in stopwords.words('english')]
    token_lem = [lem.lemmatize(word) for word in token]
    count = Counter(token_lem)
    return count
    
```

```python
count4 = PowerFunction(articles.loc[0,'text'], lem)
count4
```

```python
count4.most_common(2)
```

```python
tokens4 = [word_tokenize(document.lower()) for document in articles['text']]
```

```python
dicted = Dictionary(tokens4)
dicted
```

```python
dicted.token2id
```

```python
np.shape(dicted)
```

```python
corpus = [dicted.doc2bow(doc) for doc in tokens4]
corpus
```

```python
sort= sorted(corpus[9], key= lambda w : w[1], reverse=True)
```

```python
for word, word_count in sort[:5]:
    print(dicted.get(word), word_count)
    
```

```python
from collections import defaultdict
import itertools
```

```python
total_word_count= defaultdict(int)

for word_id, word_count in itertools.chain.from_iterable(corpus):
    total_word_count[word_id] += word_count
```

```python
sorted_word_count = sorted(total_word_count.items(), key= lambda w : w[1], reverse=True)
```

```python
for key, count in sorted_word_count[:5]:
    print(dicted.get(key), count)
```
